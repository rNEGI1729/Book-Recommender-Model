```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv("Books.csv")
ratings = pd.read_csv("Ratings.csv")
users = pd.read_csv("Users.csv")
```

# Popularity Based Recommender System

```python
# I will list top 50 Books with highest rating and atleast reviewed by 250 people.
```

```python
# Merge Ratings and Books dataset on ISBN.

ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
ratings_with_name = ratings_with_name.dropna(subset=['Book-Rating'])
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>250].sort_values('avg_rating',ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','num_ratings','avg_rating','Book-Author','Image-URL-M']]
```

```python
popular_df
```

# Collaborative filtering based Recommendation System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
educated_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>= 50
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index ='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace= True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```
